```python
import os
from datetime import datetime
path='C:/Users/user/Desktop/Python/Python基礎與物件導向概念介紹/'
print(list(os.walk(path)))
```

```python
for entry in os.scandir(path):
    info=entry.stat()
    da=datetime.utcfromtimestamp(info.st_mtime)
    dstr=da.strftime('%Y/%m/%d')
    if entry.is_dir():
        print('資料夾:',entry.name,'最後存取時間:',dstr)
    elif entry.is_file():
        print('檔案:',entry.name,'最後存取時間:',dstr)     
```

```python
for entry in os.scandir(path):
    info=entry.stat()
    da=datetime.utcfromtimestamp(info.st_mtime)
    dstr=da.strftime('%Y/%m/%d')
    if entry.is_dir():
        print(entry.name,'File folder',dstr)
    elif entry.is_file():
        ffname=os.path.join(path,entry.name)
        fsize=int(os.path.getsize(ffname)/1024)
        ext=os.path.splitext(ffname)[-1].replace('.','')
        print(entry.name+'\t\t\t',str(fsize)+'KB',ext,dstr)
```

```python

infname = 'C:/Users/user/Desktop/Python/Python基礎與物件導向概念介紹/RnJ.txt'
infile = open(infname, 'rt')
outfname = 'C:/Users/user/Desktop/Python/Python基礎與物件導向概念介紹/RnJ.out'
outfile = open(outfname, 'wt')

for line in infile:
    str1 = line.strip()
    if len(str1)>0:
        #print(str1)
        outfile.write(str1+'\n')

infile.close()
outfile.close()
```
