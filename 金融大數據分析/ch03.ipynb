```python

import matplotlib.pyplot as plt

bound = 10
xs = [i for i in range(bound+1)]
y1s = [i**2 for i in range(bound+1)]
y2s = [i**3 for i in range(bound+1)]

xt = [i for i in range(bound+1)]
yt = [i for i in range(bound**3+1) if i%100==0]

#plt.plot(xs, y1s, xs, y2s)
plt.plot(xs, y1s, 'r:o', label='SQUARE')
plt.plot(xs, y2s, 'g--s', label='CUBIC')
plt.xlabel('Base Number')
plt.ylabel('Power Result')
plt.title('Power Series')

plt.xticks(xt)
plt.yticks(yt)

plt.legend(loc='best')
plt.savefig('a.png', dpi=300)

plt.show()
```

```python

import matplotlib.pyplot as plt

bound = 6
xs = [i for i in range(bound+1)]
y1s = [i**2 for i in range(bound+1)]
y2s = [i**3 for i in range(bound+1)]

xt = [i for i in range(bound+1)]
yt = [i for i in range(bound**3+1) if i%20==0]
xxt = ['d'+str(i) for i in range(bound+1)]
print(xxt)

width = 0.4
x1s = [x-width/2 for x in xs]
x2s = [x+width/2 for x in xs]
plt.bar(x1s, y1s, width, label='SQUARE')
plt.bar(x2s, y2s, width, label='CUBIC')

plt.xlabel('Base Number')
plt.ylabel('Power Result')
plt.title('Power Series')

plt.xticks(xt, xxt)
plt.yticks(yt)

plt.legend(loc='best')
plt.savefig('a.png', dpi=300)

plt.show()
```

```python

import sqlite3
import matplotlib.pyplot as plt

# DB part
dbname = 'mystock.db'
conn = sqlite3.connect(dbname)
cur = conn.cursor()

dlist = []
cllist = []
vollist = []
sqlstr = 'SELECT date, cl, vol FROM stktrade WHERE id="^IXIC" AND date>="2020-01-01" AND date<"2020-07-01" ORDER BY date ASC'
results = cur.execute(sqlstr)
for rec in results:
    dlist.append(rec[0])
    cllist.append(rec[1])
    vollist.append(rec[2])

conn.commit()
conn.close()
print('Data length:', len(dlist))

# Fig part

xs = [i for i in range(len(dlist))]

xxt = [dlist[i] if i%5==0 else '' for i in range(len(dlist))]
#for i in range(len(dlist)):
#    if i%5==0:
#        xxt.append(dlist[i])
#    else:
#        xxt.append('')

plt.figure(figsize=(16, 5))
plt.plot(xs, cllist, label='Close')
plt.xlabel('#')
plt.ylabel('Close')
plt.xticks(xs, xxt, rotation=50)
plt.title('NASDAQ 2020')
plt.legend(loc='best')
plt.show()

plt.figure(figsize=(16, 5))
plt.bar(xs, vollist, 0.5, label='Volume')
plt.xlabel('#')
plt.ylabel('Volume')
plt.xticks(xs, xxt, rotation=50)
plt.title('NASDAQ 2020')
plt.legend(loc='best')
plt.show()

```
