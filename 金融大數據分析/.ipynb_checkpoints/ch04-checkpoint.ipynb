```python

import sqlite3

dbname = 'mystock.db'
conn = sqlite3.connect(dbname)
cur = conn.cursor()

stkid = '2303'
dstr1 = '2021-01-01'
dstr2 = '2021-09-01'
sqlstr = 'SELECT date, cl, vol FROM stktrade WHERE id="'+stkid+'" AND date>="'+dstr1+'" AND date<"'+dstr2+'" ORDER BY date ASC'

dlist = []
cllist = []
vollist = []
results = cur.execute(sqlstr)
count = 0
for rec in results:
    dlist.append(rec[0])
    cllist.append(rec[1])
    vollist.append(rec[2])
    count += 1

conn.commit()
conn.close()

print(len(dlist), 'records')
```

```python

def calcMA(series, n):
    mas = [None]*len(series)
    for i in range(n-1, len(series)):
        sublist = []
        for j in range(n):
            sublist.append(series[i-j])
        mas[i] = sum(sublist)/len(sublist) if len(sublist)>0 else 0
    return mas

ma5s = calcMA(cllist, 5)
ma10s = calcMA(cllist, 10)
ma60s = calcMA(cllist, 60)
mv5s = calcMA(vollist, 5)
mv10s = calcMA(vollist, 10)
```

```python

import matplotlib.pyplot as plt

xs = [i for i in range(len(dlist))]
xxt = [dlist[i] if i%5==0 else '' for i in range(len(dlist))]

plt.figure(figsize=(16, 5))
plt.plot(xs, cllist, 'b-,', label='Close')
plt.plot(xs, ma5s, 'r:,', label='MA5')
plt.plot(xs, ma60s, 'g:,', label='MA60')
plt.xlabel('Date')
plt.ylabel('Close')
plt.xticks(xs, xxt, rotation=50)
plt.title('2303 202101~202108')
plt.legend(loc='best')
plt.grid(color='gray', linestyle='dotted', linewidth=1)
plt.savefig('2303.202101-202108.png', dpi=120)
plt.show()
```

```python

import sqlite3

dbname = 'mystock.db'
conn = sqlite3.connect(dbname)
cur = conn.cursor()

stkid = '2454'
dstr1 = '2021-03-01'
dstr2 = '2021-09-01'
sqlstr = 'SELECT date, cl FROM stktrade WHERE id="'+stkid+'" AND date>="'+dstr1+'" AND date<"'+dstr2+'" ORDER BY date ASC'

dlist = []
cllist = []
results = cur.execute(sqlstr)
count = 0
for rec in results:
    dlist.append(rec[0])
    cllist.append(rec[1])
    count += 1

conn.commit()
conn.close()

print(len(dlist), 'records')

```

```python

def calcRSI(series, n):
    rsis = [None]*len(series)
    for i in range(n, len(series)):
        ups = []
        dns = []
        for j in range(n):
            if series[i-j]>series[i-j-1]:
                ups.append(series[i-j]-series[i-j-1])
            if series[i-j]<series[i-j-1]:
                dns.append(series[i-j-1]-series[i-j])
        avgUp = sum(ups)/len(ups) if len(ups)>0 else 0.0
        avgDn = sum(dns)/len(dns) if len(dns)>0 else 0.0
        rsis[i] = 100*avgUp/(avgUp+avgDn) if avgUp+avgDn!=0 else 0.0
    return rsis

rsi9s = calcRSI(cllist, 9)
rsi14s = calcRSI(cllist, 14)
```

```python

import matplotlib.pyplot as plt

xs = [i for i in range(len(dlist))]
xxt = [dlist[i] if i%5==0 else '' for i in range(len(dlist))]

plt.figure(figsize=(16, 5))
plt.plot(xs, rsi9s, 'b:,', label='RSI9')
plt.plot(xs, rsi14s, 'r:,', label='RSI14')
plt.xlabel('Date')
plt.ylabel('RSI(%)')
plt.xticks(xs, xxt, rotation=50)
plt.title('2454 202103~202108')
plt.grid(color='gray', linestyle='dotted', linewidth=1)
plt.legend(loc='best')
plt.show()
```
