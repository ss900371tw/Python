```python

import requests
import time
import random

urlprefix = 'https://www.twse.com.tw/exchangeReport/MI_INDEX?response=csv&type=ALL&date='
headerstr = {'User-Agent':'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/51.0.2704.103 Safari/537.36'}

infile = open('date2021.lst', 'rt', encoding='utf-8')
index = 1
for line in infile:
    dstr = line.strip()
    urlstr = urlprefix+dstr
    #print(urlstr)
    try:
        resp = requests.get(urlstr, headers=headerstr)
        resp.raise_for_status()
        outfile = open(dstr+'.csv', 'wt', encoding='utf-8')
        outfile.write(resp.text)
        outfile.close()
        print(index, dstr+'.csv saved.')
    except:
        print('Error fetching date:', dstr)
    index += 1
    time.sleep(random.randint(5, 7))
    
infile.close()
```

```python

def processDate(dstr):
    infname = dstr+'.csv'
    infile = open(infname, 'rt', encoding='utf-8')
    csvReader = csv.reader(infile)
    inData = False
    alldata = []
    for row in csvReader: 
        if len(row)>0 and row[0]=='':
            inData = True
            continue
        if inData:
            if len(row)>0 and row[0].startswith('='):
                continue
            else:
                data = []
                try:
                    stkid = str(int(row[0]))
                    data.append(stkid)
                    data.append(row[1])
                    for i in range(2, 4+1):
                        data.append(int(row[i].replace(',', '')))
                    for i in range(5, 8+1):
                        data.append(float(row[i].replace(',', '')))
                    alldata.append(data)
                except:
                    pass
    infile.close()
    return alldata

infile = open('date2021.lst', 'rt', encoding='utf-8')
index = 1
market2021 = {}
count = 0
for line in infile:
    dstr = line.strip()
    print('Process ', dstr)
    alldata = processDate(dstr)
    market2021[dstr] = alldata
    count += len(alldata)
    
infile.close()

print('# of records', count)
```

```python

#market2021 dstr => list of list(stkid, stkname, vol, tra, val, op, hi, lo, cl)
stkids = {}

for dstr, alldata in market2021.items():
    for data in alldata:
        stkids[data[0]] = data[1]
        
print('total', len(stkids.keys()), 'unique stocks')
for k, v in stkids.items():
    print(k, '=>', v)
```

```python

import sqlite3

#market2021 dstr => list of list(stkid, stkname, vol, tra, val, op, hi, lo, cl)
dbname = 'mystock.db'
conn = sqlite3.connect(dbname)

sqlstr = 'CREATE TABLE stkinfo (id TEXT, name TEXT, PRIMARY KEY(id));'
conn.execute(sqlstr)
print('TABLE stkinfo created.')

sqlstr = 'CREATE TABLE stktrade (id TEXT, date TEXT, op REAL, hi REAL, lo REAL, \
            cl REAL, vol INTEGER, tra INTEGER, val INTEGER, PRIMARY KEY(id, date));'
conn.execute(sqlstr)
print('TABLE stktrade created.')

conn.commit()
conn.close()
```

```python

import sqlite3

dbname = 'mystock.db'
conn = sqlite3.connect(dbname)

count = 0
for stkid, stkname in stkids.items():
    sqlstr = 'INSERT INTO stkinfo (id, name) VALUES ("'+stkid+'", "'+stkname+'");'
    #print(sqlstr)
    conn.execute(sqlstr)
    count += 1
    
print('total', count, 'records inserted.')

conn.commit()
conn.close()

```

```python

import sqlite3

#market2021 dstr => list of list(stkid, stkname, vol, tra, val, op, hi, lo, cl)

def makeSQLString(dstr, data):
    # 20210616 => SQL: 2021-06-16
    dstr2 = dstr[:4]+'-'+dstr[4:6]+'-'+dstr[6:]
    sqlstr = 'INSERT INTO stktrade (id, date, op, hi, lo, cl, vol, tra, val) VALUES '
    sqlstr += '("'+data[0]+'", "'+dstr2+'", '+str(data[5])+', '+str(data[6])+', '+str(data[7])+', '+str(data[8])+', '
    sqlstr += str(data[2])+', '+str(data[3])+', '+str(data[4])+');'
    return sqlstr

dbname = 'mystock.db'
conn = sqlite3.connect(dbname)

count = 0
for dstr in market2021:
    alldata = market2021[dstr]
    for data in alldata:
        sqlstr = makeSQLString(dstr, data)
        #print(sqlstr)
        conn.execute(sqlstr)
        count += 1
        if count%10000==0:
            print(count, 'records inserted.')
    conn.commit()
conn.close()

print('total', count, 'records inserted.')
```
