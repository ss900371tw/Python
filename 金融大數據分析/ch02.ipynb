```python
# https://query1.finance.yahoo.com/v7/finance/download/%5EIXIC?
#period1=34560000&
#period2=1630540800&
#interval=1d&
#events=history&
#includeAdjustedClose=true

from datetime import date, datetime
import requests
import time
import random

def makeTimeStamp(dstr):
    # 2021-09-01
    slist = dstr.split('-')
    return int(datetime(int(slist[0]), int(slist[1]), int(slist[2]), 0, 0).timestamp())+28800

def makeYahooFinanceURL(stkid, dstr1, dstr2):
    urlpart1 = 'https://query1.finance.yahoo.com/v7/finance/download/'
    urlpart2 = '?interval=1d&events=history&includeAdjustedClose=true&'
    period1 = makeTimeStamp(dstr1)
    period2 = makeTimeStamp(dstr2)
    #print(period1, period2)
    urlstr = urlpart1+stkid+urlpart2+'period1='+str(period1)+'&period2='+str(period2)
    return urlstr

headerstr = {'User-Agent':'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/51.0.2704.103 Safari/537.36'}

stkids = {'^IXIC':'NASDAQ', 'AAPL':'Apple', 'AMZN':'Amazon', 'FB':'Facebook', 
          'NFLX':'Netflix', 'GOOG':'Google', 'NVDA':'NVidia', 'TSLA':'Telsa'}
dstr1 = '2012-01-01'
dstr2 = '2021-09-01'
for stkid, stkname in stkids.items():
    urlstr = makeYahooFinanceURL(stkid, dstr1, dstr2)
    #print(stkname, '=>', urlstr)
    try:
        resp = requests.get(urlstr, headers=headerstr)
        resp.raise_for_status()
        outfname = stkid+'.'+dstr1+'.'+dstr2+'.csv'
        with open(outfname, 'wt', encoding='utf-8') as outfile:
            outfile.write(resp.text)
        print(outfname, 'saved.')
    except:
        print('Fetch', stkid, 'failed.')
        pass
    time.sleep(random.randint(5, 7))
```

```python

import csv

stkids = {'^IXIC':'NASDAQ', 'AAPL':'Apple', 'AMZN':'Amazon', 'FB':'Facebook', 
          'NFLX':'Netflix', 'GOOG':'Google', 'NVDA':'NVidia', 'TSLA':'Telsa'}
dstr1 = '2012-01-01'
dstr2 = '2021-09-01'

marketNASDAQ = {}
count = 0
for stkid in stkids:
    infname = stkid+'.'+dstr1+'.'+dstr2+'.csv'
    infile = open(infname, 'rt', encoding='utf-8')
    csvReader = csv.reader(infile)
    alldata = []
    for row in csvReader:
        if row[0]=='Date':
            continue
        data = [row[i] for i in range(len(row)) if i!=5]
        alldata.append(data)
    infile.close()
    marketNASDAQ[stkid] = alldata
    print(stkid, 'has', len(alldata), 'records processed.')
    count += len(alldata)

print('Total', count, 'records processed.')
```

```python

import sqlite3

def makeSQLString(stkid, data):
    sqlstr = 'INSERT INTO stktrade (id, date, op, hi, lo, cl, vol, tra, val) VALUES ('
    sqlstr += '"'+stkid+'", "'+data[0]+'", '+data[1]+', '+data[2]+', '+data[3]+', '+data[4]+', '+data[5]+', 0, 0);'
    return sqlstr

stkids = {'^IXIC':'NASDAQ', 'AAPL':'Apple', 'AMZN':'Amazon', 'FB':'Facebook', 
          'NFLX':'Netflix', 'GOOG':'Google', 'NVDA':'NVidia', 'TSLA':'Telsa'}
dbname = 'mystock.db'
conn = sqlite3.connect(dbname)

count = 0
for stkid in stkids:
    alldata = marketNASDAQ[stkid]
    for data in alldata:
        sqlstr = makeSQLString(stkid, data)
        conn.execute(sqlstr)
        count += 1
        if count%1000==0:
            print(count, 'records inserted.')
    conn.commit()
    
print('Total', count, 'records inserted.')
conn.close()
```

```python

import sqlite3

stkid = '^IXIC'
dstr1 = '2021-02-01'
dstr2 = '2021-07-01'

dbname = 'mystock.db'
conn = sqlite3.connect(dbname)
cur = conn.cursor()

sqlstr = "SELECT date, cl FROM stktrade WHERE id='"+stkid+"' AND date>='"+dstr1+"' AND date<'"+dstr2+"' ORDER BY date ASC;"
#print(sqlstr)
results = cur.execute(sqlstr)
count = 0
outfname = stkid+'.'+dstr1+'.'+dstr2+'.csv'
outfile = open(outfname, 'wt', encoding='utf-8')
outfile.write('DATE,PRICE\n')
for rec in results:
    #print(rec)
    line = rec[0]+','+str(rec[1])+'\n'
    outfile.write(line)
    count += 1

print(count, 'records')

outfile.close()
conn.commit()
conn.close()

```

```python

import sqlite3

stkid = 'GOOG'
dstr1 = '2021-01-01'
dstr2 = '2021-09-01'

dbname = 'mystock.db'
conn = sqlite3.connect(dbname)
cur = conn.cursor()

sqlstr = "SELECT date, hi, lo, cl FROM stktrade WHERE id='"+stkid+"' AND date>='"+dstr1+"' AND date<'"+dstr2+"' ORDER BY date ASC;"
#print(sqlstr)
results = cur.execute(sqlstr)
count = 0
outfname = stkid+'.'+dstr1+'.'+dstr2+'.csv'
outfile = open(outfname, 'wt', encoding='utf-8')
outfile.write('DATE,XXX\n')
prevCL = 0
for rec in results:
    #print(rec)
    #line = rec[0]+','+str(rec[1])+','+str(rec[2])+'\n'
    x = 100*(rec[1]-rec[2])/prevCL if prevCL!=0 else 0
    line = rec[0]+','+str(x)+'\n'
    outfile.write(line)
    prevCL = rec[3]
    count += 1

print(count, 'records')

outfile.close()
conn.commit()
conn.close()
```
