```python

list1 = [10, 20, 30, 40, 50, 60, 70, 80, 90]
list2 = [15, 25, 35, 45, 55, 65, 75, 85, 95]
list3 = [list1[i]*3.1416+list2[i]*2.71828 for i in range(len(list1))]
print(list3)
```

```python

import numpy as np

inSeq1 = np.array(list1)
inSeq2 = np.array(list2)
outSeq = np.array(list3)
inSeq1 = inSeq1.reshape(len(list1), 1)
inSeq2 = inSeq2.reshape(len(list2), 1)
outSeq = outSeq.reshape(len(list3), 1)

dataset = np.hstack((inSeq1, inSeq2, outSeq))
print(dataset)
```

```python

def splitSequence(seq, steps):
    x = []
    y = []
    len1 = len(seq)
    for i in range(len1):
        end_ix = i+steps
        if end_ix>len1:
            break
        x.append(seq[i:end_ix, :-1])
        y.append(seq[end_ix-1, -1])
    return x, y

n_steps = 3
listX, listY = splitSequence(dataset, n_steps)
for x, y in zip(listX, listY):
    print(x, y)
    
arrayX = np.array(listX)
arrayY = np.array(listY)
print(arrayX.shape)
print(arrayY.shape)
```

```python

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv1D, MaxPooling1D, Flatten, Dense
from tensorflow.keras.optimizers import Adam

model = Sequential()

model.add(Conv1D(32, 2, activation='relu', input_shape=(arrayX.shape[1], arrayX.shape[2])))
model.add(MaxPooling1D(pool_size=2))
model.add(Flatten())
model.add(Dense(16, activation='relu'))
model.add(Dense(1))

print(model.summary())
model.compile(loss='mse', optimizer=Adam(lr=0.001))
```

```python

import matplotlib.pyplot as plt

train_history = model.fit(arrayX, arrayY, epochs=4000)

plt.plot(train_history.history['loss'], label='Loss')
plt.xlabel('Epoch')
plt.ylabel('MSE')

plt.show()
```

```python
p1 = 3.1416
p2 = 2.71828

x_input = np.array([[60, 65], [70, 75], [80, 85]])
x_input = x_input.reshape(1, 3, 2)
yhat = model.predict(x_input)
print(yhat, 80*p1+85*p2)

x_input = np.array([[80, 85], [90, 95], [100, 105]])
x_input = x_input.reshape(1, 3, 2)
yhat = model.predict(x_input)
print(yhat, 100*p1+105*p2)

x_input = np.array([[180, 185], [190, 195], [200, 205]])
x_input = x_input.reshape(1, 3, 2)
yhat = model.predict(x_input)
print(yhat, 200*p1+205*p2)

x_input = np.array([[283, 288], [293, 298], [303, 308]])
x_input = x_input.reshape(1, 3, 2)
yhat = model.predict(x_input)
print(yhat, 303*p1+308*p2)


```
