```python

def splitSequence(seq, steps_in, steps_out):
    x = []
    y = []
    len1 = len(seq)
    for i in range(len1):
        end_ix = i+steps_in
        out_end_ix = end_ix+steps_out
        if out_end_ix>len1:
            break
        x.append(seq[i:end_ix])
        y.append(seq[end_ix:out_end_ix])
    return x, y

list1 = []
for i in range(100):
    list1.append((i+1)*10+5)

n_steps_in = 5
n_steps_out = 3
listX, listY = splitSequence(list1, n_steps_in, n_steps_out)
for x, y in zip(listX, listY):
    print(x, y)
  
```

```python

import numpy as np

arrayX = np.array(listX)
arrayY = np.array(listY)
print(arrayX.shape)
print(arrayY.shape)

arrayX = arrayX.reshape(arrayX.shape[0], arrayX.shape[1], 1)
print(arrayX.shape)
```

```python

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv1D, MaxPooling1D, Flatten, Dense
from tensorflow.keras.optimizers import Adam

model = Sequential()

model.add(Conv1D(32, 2, activation='relu', input_shape=(arrayX.shape[1], 1)))
model.add(MaxPooling1D(pool_size=2))
model.add(Flatten())
model.add(Dense(16, activation='relu'))
model.add(Dense(n_steps_out))

print(model.summary())
model.compile(loss='mse', optimizer=Adam(lr=0.001))
```

```python

import matplotlib.pyplot as plt

train_history = model.fit(arrayX, arrayY, epochs=2500)

plt.plot(train_history.history['loss'], label='Loss')
plt.xlabel('Epoch')
plt.ylabel('MSE')

plt.show()
```
