```python

import sqlite3

dbname = 'mystock.db'
conn = sqlite3.connect(dbname)
cur = conn.cursor()
cllist = []
oplist = []
sqlstr = 'SELECT date, cl, op FROM stktrade WHERE id="2330" AND date>="2021-01-01" AND date<"2021-09-01"'
results = cur.execute(sqlstr)
for rec in results:
    cllist.append(rec[1])
    oplist.append(rec[2])
conn.commit()
conn.close()

print(len(cllist))
print(cllist)
```

```python

n_steps = 7
list1 = cllist[:-1]
list2 = oplist[:-1]
list3 = cllist[1:]
print(len(list1), len(list2), len(list3))

```

```python

import numpy as np

inSeq1 = np.array(list1)
inSeq2 = np.array(list2)
outSeq = np.array(list3)
inSeq1 = inSeq1.reshape(len(list1), 1)
inSeq2 = inSeq2.reshape(len(list2), 1)
outSeq = outSeq.reshape(len(list3), 1)

dataset = np.hstack((inSeq1, inSeq2, outSeq))
print(dataset)
```

```python


def splitSequence(seq, steps):
    x = []
    y = []
    len1 = len(seq)
    for i in range(len1):
        end_ix = i+steps
        if end_ix>len1:
            break
        x.append(seq[i:end_ix, :-1])
        y.append(seq[end_ix-1, -1])
    return x, y

n_steps = 7
listX, listY = splitSequence(dataset, n_steps)
for x, y in zip(listX, listY):
    print(x, y)
    
arrayX = np.array(listX)
arrayY = np.array(listY)
print(arrayX.shape)
print(arrayY.shape)

```

```python

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv1D, MaxPooling1D, Flatten, Dense
from tensorflow.keras.optimizers import Adam

model = Sequential()

model.add(Conv1D(32, 2, activation='relu', input_shape=(arrayX.shape[1], arrayX.shape[2])))
model.add(MaxPooling1D(pool_size=2))
model.add(Flatten())
model.add(Dense(16, activation='relu'))
model.add(Dense(1))

print(model.summary())
model.compile(loss='mse', optimizer=Adam(lr=0.001))
```

```python

import matplotlib.pyplot as plt

train_history = model.fit(arrayX, arrayY, epochs=4000)

plt.plot(train_history.history['loss'], label='Loss')
plt.xlabel('Epoch')
plt.ylabel('MSE')

plt.show()
```
