```python

import sqlite3

dbname = 'mystock.db'
conn = sqlite3.connect(dbname)
cur = conn.cursor()
cllist = []

```

```python
sqlstr = 'SELECT date, cl FROM stktrade WHERE id="2330" AND date>="2021-01-01" AND date<"2021-09-01"'
results = cur.execute(sqlstr)
for rec in results:
    cllist.append(rec[1])
conn.commit()
conn.close()

print(len(cllist))
print(cllist)
```

```python

train_data = cllist[:-10]
test_data = cllist[-10:]
print(test_data)
```

```python

def splitSequence(seq, steps):
    x = []
    y = []
    len1 = len(seq)
    for i in range(len1):
        end_ix = i+steps
        if end_ix>len1-1:
            break
        x.append(seq[i:end_ix])
        y.append(seq[end_ix])
    return x, y

n_steps = 9
listX, listY = splitSequence(train_data, n_steps)
for x, y in zip(listX, listY):
    print(x, y)
```

```python

import numpy as np

arrayX = np.array(listX)
arrayY = np.array(listY)
print(arrayX.shape)
print(arrayY.shape)

arrayX = arrayX.reshape(arrayX.shape[0], arrayX.shape[1], 1)
print(arrayX.shape)
```

```python

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv1D, MaxPooling1D, Flatten, Dense
from tensorflow.keras.optimizers import Adam

model = Sequential()

model.add(Conv1D(64, 2, activation='relu', input_shape=(arrayX.shape[1], 1)))
model.add(MaxPooling1D(pool_size=2))
model.add(Flatten())
model.add(Dense(32, activation='relu'))
model.add(Dense(1))

print(model.summary())
model.compile(loss='mse', optimizer=Adam(lr=0.001))
```

```python

import matplotlib.pyplot as plt

train_history = model.fit(arrayX, arrayY, epochs=3000)

plt.plot(train_history.history['loss'], label='Loss')
plt.xlabel('Epoch')
plt.ylabel('MSE')

plt.show()
```
